```python
y=filtered_df['price']
x=filtered_df.drop(columns=['price','date','street','city','statezip','country'])
x.shape
```

```python
city_encoded=pd.get_dummies(filtered_df['city'],prefix='city')
statezip_encoded=pd.get_dummies(filtered_df['statezip'],prefix='statezip')
x_new=pd.concat([x,city_encoded,statezip_encoded],axis=1)
x_new.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_new,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
type(model)
```

```python
predictions=model.predict(x_test)
predictions[:5]
```

```python
y_test.head()
```

```python
r2=r2_score(y_test,predictions)
mse=mean_squared_error(y_test,predictions)
print("R2:",r2)
print("MSE:",mse)
```

```python
import joblib
joblib.dump(model,"house_price_model.pkl")
joblib.dump(city_encoded.columns, "../models/city_columns.pkl")
joblib.dump(statezip_encoded.columns, "../models/statezip_columns.pkl")
```
